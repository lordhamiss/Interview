## Approaches for deviding the problem into subproblems:

### 1- Bottom-Up Approach
The key=buidling the solution for one case off of the previous case
* We know how to solve the problem for a simple case == like one element 
* Then we figure out how to solve the proble for two cases and go on to n

### 2- Top-Down Approach
Think about how to divide the problem for case N into subproblems 


### 3- Half-and-Half Approach
The key=devide the data set in half

* Binary Search --> looking for an element in a sorted array --> first figure out which half contains the element --> then do it recursively
* Merge Sort --> sort two halfes --> merge them

### Dynamic Programming 

take in a recursive algorithm --> find the overlap subproblems --> solve them --> cache them --> use them in future

Time Complexity --> draw the recursive calls as a tree --> count the nodes

## Fibonatchi

```python
##fibonatchi recursive
def fibRec(n):
    if n==0 or n==1: 
        return 1
    return fibRec(n-1)+fibRec(n-2)
```

```python
print (fibRec(5))
```

#### Time complexity


Time complexity = O(2^n)
better insight --> the right subtrees are shorter than left subtrees --> O(1.6^n)

```python
##fibonatchi DP
def fibDP(n, cache=None):
    if cache is None:
        cache = [0 for i in range(n+1)]
    
    if n==0 or n==1:
        return 1
    if cache[n]==0:
        cache[n]==fibDP(n-2, cache) + fibDP(n-1, cache)

    return cache[n]
```

```python
##fibonatchi DP
def fibDP(n, cache=None):
    if cache is None:
        cache = []
    for i in range(n):
        if i==0 or i==1:
            cache.append(1)
        else:
            cache.append(cache[i-1]+cache[i-2])
    return cache[n-1]+cache[n-2]
```

```python
print (fibDP(2))
```
