## Definition

linked list is a data structure that represents a sequence of nodes. In a singly linked list, each node 
points to the next node in the linked list. 

A doubly linked list gives each node pointers to both the next node and the previous node.



Access to kth index --> requires k iterations --> unlike the arrays 

Adding and removing from the head --> constant time 

**Remember** --> sometimes link list problems will be solved using recursive approaches
